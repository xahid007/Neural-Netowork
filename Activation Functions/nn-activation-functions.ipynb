```python
import numpy as np
import matplotlib.pyplot as plt
```

# Activation Functions

```python
x = np.arange(-10,10,.1)
```

```python
#Sigmoid
def sigmoid(x):
    sigmoid = 1/ (1 + np.exp(-x))
    plt.plot(x,sigmoid)
    plt.axvline(x = 0, color = 'black', linestyle = '--')
sigmoid(x)
```

```python
#tanh
def Tanh(x):
    tanh = (np.exp(x)- np.exp(-x))/(np.exp(x)+ np.exp(-x))
    plt.plot(x,tanh, label= 'tanh')
    plt.axvline(x = 0, color = 'black', linestyle = '--')
    plt.axhline(y = 0, color = 'black', linestyle = '--')
    plt.legend()
    plt.plot()
Tanh(x)
```

```python
plt.plot(x, np.tanh(x)) #anotehr way to do the same thing
```

```python
#ReLU

def rectified(x):
    return max(0.0, x)

X = [x for x in range(-10, 11)]
r = [rectified(x) for x in X]
plt.plot(X, r)
```

```python
# Variations of ReLU

#leaky ReLU:
def Lrectified(x,alpha):
    return max(alpha*x,x)

X = [x for x in range(-10, 10)]
r = [Lrectified(x,.2) for x in X]

plt.xlim((-10,10))
plt.ylim((-10,10))

plt.plot([0,0],[-10,10],color='red')
plt.plot([-10,10],[0,0],color='red')
plt.plot(X, r)
```

```python
#ELU

def ELU(alpha,x):
    if (x<0):
        return alpha*(np.exp(x) - 1)
    else:
        return x

X = [x for x in range(-10, 10)]
r = [ELU(.6,x) for x in X]

plt.xlim((-10,10))
plt.ylim((-10,10))

plt.plot([0,0],[-10,10],color='red')
plt.plot([-10,10],[0,0],color='red')
plt.plot(X, r)
```
